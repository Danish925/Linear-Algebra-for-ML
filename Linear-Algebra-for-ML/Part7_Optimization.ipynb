## 1. Why Optimization is Needed

In real-world data:
- Exact solutions rarely exist
- Data is noisy
- Systems are overdetermined

ðŸ‘‰ ML models find the **best approximate solution**.


````python

import numpy as np
import matplotlib.pyplot as plt
````


## 2. Loss Function

A **loss function** measures how wrong a model is.

Example (Mean Squared Error):
```
L = (yÌ‚ âˆ’ y)Â²
```

ML goal:
> Minimize the loss


````python

# Simple loss example
y_true = np.array([2, 4, 6])
y_pred = np.array([2.2, 3.8, 5.5])

loss = np.mean((y_pred - y_true) ** 2)
print("Loss:", loss)
````


## 3. Gradient (Core Idea)

The **gradient** tells:
> Direction of steepest increase of loss

To reduce loss, we move in the **opposite direction**.



## 4. Gradient Descent

Update rule:
```
w = w âˆ’ Î· Â· âˆ‡L
```

Where:
- `Î·` â†’ learning rate
- `âˆ‡L` â†’ gradient


````python

# Gradient descent for y = w*x

x = np.array([1, 2, 3, 4])
y = np.array([2, 4, 6, 8])

w = 0.0
lr = 0.1

for i in range(10):
    y_pred = w * x
    grad = np.mean(2 * (y_pred - y) * x)
    w = w - lr * grad
    print(f"Iteration {i}: w = {w}")
````


## 5. Learning Rate

Learning rate controls **step size**.

- Too large â†’ divergence
- Too small â†’ slow learning



## 6. Visualizing Loss Curve


````python

# Visualizing loss over iterations
w = 0.0
lr = 0.1
losses = []

for i in range(30):
    y_pred = w * x
    loss = np.mean((y_pred - y) ** 2)
    losses.append(loss)
    grad = np.mean(2 * (y_pred - y) * x)
    w = w - lr * grad

plt.plot(losses)
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.title("Loss decreases with Gradient Descent")
plt.show()
````


## 7. Gradient is a Vector

For multiple parameters:
- Gradient has **one component per parameter**

This links directly to **vectors and matrices**.



## 8. Why Gradient Descent Works

- Efficient for large datasets
- Scales well with dimensions
